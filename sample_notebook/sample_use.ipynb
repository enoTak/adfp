```python
from adfp import Variable, plot_dot_graph
import numpy as np
```

```python
def sphere(x, y):
    z = x ** 2 + y ** 2
    return z
```

```python
x = Variable(np.array(1.0))
y = Variable(np.array(2.0))
z = sphere(x, y)
z.backward()
print(x.grad, y.grad)
```

```python
def matyas(x, y):
    z = 0.26 * (x ** 2 + y ** 2) - 0.48 * x * y
    return z
```

```python
x = Variable(np.array(1.0))
y = Variable(np.array(1.0))
z = matyas(x, y)
z.backward()
print(x.grad, y.grad)
```

```python
def goldstein(x, y):
    z = (1 + (x + y + 1)**2 * (19 - 14*x + 3*x**2 - 14*y + 6*x*y + 3*y**2)) * \
        (30 + (2*x - 3*y)**2 * (18 - 32*x + 12*x**2 + 48*y - 36*x*y + 27*y**2))
    return z
```

```python
x = Variable(np.array(1.0), name='x')
y = Variable(np.array(1.0), name='y')
z = goldstein(x, y)
z.name = 'z'
z.backward()
print(x.grad, y.grad)
```

```python
plot_dot_graph(z, verbose=False, to_file="goldstein.png")
```
