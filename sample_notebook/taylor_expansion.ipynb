```python
from pyautodiff import *
import numpy as np
import math
```

```python
def sin_taylor(x, threshold=0.0001):
    y = 0
    for i in range(100000):
        c = (-1) ** i / math.factorial(2 * i + 1)
        t = c * x ** (2 * i + 1)
        y = y + t
        if (abs(t.data) < threshold):
            break
    return y
```

```python
x = Variable(np.array(np.pi/4))
y = sin_taylor(x)
y.backward()

print(y.data, x.grad)
```

```python
plot_dot_graph(y, verbose=False)
```
